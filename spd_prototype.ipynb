```python
#Translation with voice
from tkinter import *
from tkinter import ttk
from deep_translator import GoogleTranslator
import speech_recognition as sr

# Initialize the main application window
root = Tk()
root.title("Language Translation App")
root.geometry("500x400")

# Define the list of supported languages
languages = GoogleTranslator().get_supported_languages()
language_codes = GoogleTranslator().get_supported_languages(as_dict=True)

# Function to perform translation
def translate_text():
    source_lang = language_codes[source_lang_combobox.get()]
    target_lang = language_codes[target_lang_combobox.get()]
    text = source_text.get("1.0", "end-1c")  # Fetch text from text box

    if text:
        translator = GoogleTranslator(source=source_lang, target=target_lang)
        translated_text = translator.translate(text)
        target_text.delete("1.0", "end")  # Clear previous translation
        target_text.insert("1.0", translated_text)  # Display new translation

# Function to handle voice input
def voice_to_text():
    recognizer = sr.Recognizer()
    with sr.Microphone() as source:
        target_text.delete("1.0", "end")  # Clear any previous text
        target_text.insert("1.0", "Listening...")  # Indicate listening state
        try:
            audio = recognizer.listen(source)
            spoken_text = recognizer.recognize_google(audio)
            source_text.delete("1.0", "end")  # Clear previous input
            source_text.insert("1.0", spoken_text)  # Insert recognized text
            target_text.delete("1.0", "end")  # Clear listening indication
        except sr.UnknownValueError:
            target_text.delete("1.0", "end")
            target_text.insert("1.0", "Sorry, I couldn't understand the audio.")
        except sr.RequestError as e:
            target_text.delete("1.0", "end")
            target_text.insert("1.0", f"Error: {e}")

# Source Language Selection
Label(root, text="Source Language").pack(pady=5)
source_lang_combobox = ttk.Combobox(root, values=list(languages))
source_lang_combobox.set("english")  # Set default language
source_lang_combobox.pack()

# Target Language Selection
Label(root, text="Target Language").pack(pady=5)
target_lang_combobox = ttk.Combobox(root, values=list(languages))
target_lang_combobox.set("french")  # Set default language
target_lang_combobox.pack()

# Source Text Entry
Label(root, text="Enter Text").pack(pady=5)
source_text = Text(root, height=5, width=40)
source_text.pack()

# Voice Input Button
voice_button = Button(root, text="Speak", command=voice_to_text)
voice_button.pack(pady=5)

# Translate Button
translate_button = Button(root, text="Translate", command=translate_text)
translate_button.pack(pady=10)

# Target Text Display
Label(root, text="Translated Text").pack(pady=5)
target_text = Text(root, height=5, width=40)
target_text.pack()

# Run the main loop
root.mainloop()
```
